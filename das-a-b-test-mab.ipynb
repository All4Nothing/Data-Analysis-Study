### A/B Test, MAB  

**A/B Test** 
A/B 테스트는 기존 시스템을 개선하거나 새로운 시도를 할 때 성과를 측정하기 위해, 두 집단으로 나눠서 서로 다른 콘텐츠를 보여준 다음 어떤 집단이 더 높은 성과를 보이는지를 비교하는 방법이다.  
A/B 테스트를 수행할 수 있는 주제는 다음과 같은 것들이 있다.  
- 웹사이트 내비게이션 배치와 문구 등 UI/UX 최적화  
- 추천 콘텐츠 노출 리스트 최적화
- 상품 종류와 가격, 노출 사이즈 등 최적화
- 서비스 가입 경로 시나리오 최적화
- 게임 아이템 드롭률 최적화
- 배너/버튼의 위치, 사이즈, 색상, 폰트, 문구 등 최적화
- 검색엔진의 키워드 검색 결과 최적화  
A/B 테스트를 하기 위해 표본 그룹은 편향되지 않도록 무작위 추출을 올바르게 시행해야 한다. 또한, 표본 수는 일반적으로 1,000명 이상의 표본으로 테스트 하는 것이 좋다. 또한, A/B 테스트는 일시적으로 하는 것이 아니라, 지속적으로 진행하는 것이 좋다. 또한, A/B 테스트를 기획할 때 한 사용자가 한 가지 변화만 경험하도록 설계해야 한다.  
A/B 테스트의 단점은, 테스트 가능한 경우의 수가 매우 다양하기 때문에 최적의 조건을 찾을 때까지 오랜 시간과 비용이 필요하다는 것이다. 덜 효과적인 조건의 테스트에 배정되는 인원들은 계속 늘어나고, 잠재적 손실 비용은 커져 간다. 그리고 트렌드는 계속해서 변한다.  

**MAB(Multi-Armed Bandit)**  
MAB의 Multi-armed는 여러 개의 손잡이를 뜻하고, Bandit은 슬롯머신을 의미한다. 수익률이 다르고 계속 변화하는 여러 개의 슬롯 머신이 있는 카지노에서 일정 시간안에 도박꾼이 최대의 이익을 얻으러면 어떤 전략으로 슬롯머신을 플레이 해야하는지가 알고리즘의 출발점이다. Exploration은 최적의 보상을 얻을 수 있는 방안을 찾기 위해 계속해서 실험하는 것을 말하고 Exploitation은 선택의 최적의 방안을 계속 실행하여 이익을 최대화 하는 것이을 말한다. Exploration과 Exploitation은 trade-off 관계이기 때문에, 이 조합을 최적화하여 최대의 이익을 얻는 것이 MAB의 목적이다.  
대표적으로 세 가지 MAB 알고리즘이 있다.  
- ε-greedy : 일정 확률로 대안을 무작위로 탐색하는 알고리즘
- UCB : 대안의 결과가 불확실한 것을 우선순위로 탐색하는 알고리즘
- Thompson Sampling : 대안의 결과에 대한 베타 분포를 계산하여 샘플링한 값 중 가장 높은 대안을 탐색하는 알고리즘  
MAB 알고리즘은 기계학습 모델에도 적용이 가능하다. 추천 시스템 모델에서 여러 종류의 협업 필터링과 콘텐츠 기반 필터링 기법을 앙상블하여 사용한다고 했을 때, MAB를 적용하여 최적의 추천 정확도를 보이는 모델을 선택할 수 있도록 할 수 있다. 일반적인 앙상블 방식은 대부분 테스트 데이터의 정확도를 가지고 모델 결과를 가중 합하여 사용한다. 하지만 MAB를 적용하면 사용자의 실제 반응 정보를 바로 반응하여 실제 반응률이 가장 좋은 모델을 적용한 아이템 추천을 하여 반응률을 극대화할 수 있다.
