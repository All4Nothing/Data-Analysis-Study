Boxplot Structure

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# https://www.kaggle.com/datasets/karthickveerakumar/startup-logistic-regression
df = pd.read_csv("../input/startup-logistic-regression/50_Startups.csv")
```

```python
df.head()
```

### Boxplot

```python
# 세로 boxplot
plt.figure(figsize=(8,6))
sns.boxplot(y='Profit', data=df)
plt.show()

# 세로 boxplot
plt.figure(figsize=(8,2))
sns.boxplot(x='Profit', data=df)
plt.show()
```

*최솟값을 넘어가는 극단값이 존재함을 알 수 있으며, 중앙값이 약간 낮은 위치에 있는 것을 보아 데이터 분포가 오른쪽으로 치우쳐 있음을 알 수 있다.*

```python
# state 구분에 따른 profit boxplot 시각화
plt.figure(figsize=(8,5))
sns.boxplot(x='State', y='Profit', data=df)
plt.show()
```

```python
# 평균, 데이터 포인트 포함한 boxplot
sns.boxplot(x='State', y='Profit', showmeans=True, boxprops={'facecolor':'None'}, data=df)

sns.stripplot(x='State', y='Profit', data=df, jitter=True, marker='o', alpha=0.5, color='black')

plt.show()
```

*평균값 위치와 실제 데이터 포인트들을 알 수 있다.  
Florida의 평균값과 중앙값 차이가 가장 큰 것으로 보아 왼꼬리 분포의 성향이 가장 강할 것으로 판단된다.*
